```python
import numpy as np
from scipy.stats import pearsonr, spearmanr, kendalltau
import matplotlib.pyplot as plt

# Step 1: Define the correlation values from -1 to 1 in steps of 0.05
correlation_values = np.arange(-1, 1.05, 0.05)

n=100

# Step 2: Create empty arrays to store correlation coefficients
pearson_correlations_linear = []
spearman_correlations_linear = []
kendall_correlations_linear = []

pearson_correlations_exponential = []
spearman_correlations_exponential = []
kendall_correlations_exponential = []

pearson_correlations_u_shaped = []
spearman_correlations_u_shaped = []
kendall_correlations_u_shaped = []

# Step 3: Loop through correlation values and calculate the correlations for different relationships
for rho in correlation_values:
    # Linear Relationship
    np.random.seed(0)
    x_linear = np.random.rand(n)
    y_linear = rho * x_linear + (1 - rho) * np.random.rand(n)

    # Exponential Relationship
    np.random.seed(0)
    x_exp = np.random.rand(n)
    y_exp = np.exp(rho * x_exp) + 0.1 * np.random.rand(n)

    # U-shaped Relationship
    np.random.seed(0)
    x_u_shaped = np.random.rand(n)
    y_u_shaped = (rho * x_u_shaped - 0.5)**2 + 0.1 * np.random.rand(n)

    # Calculate the correlation coefficients for each relationship
    pearson_corr_linear, _ = pearsonr(x_linear, y_linear)
    spearman_corr_linear, _ = spearmanr(x_linear, y_linear)
    kendall_corr_linear, _ = kendalltau(x_linear, y_linear)

    pearson_corr_exp, _ = pearsonr(x_exp, y_exp)
    spearman_corr_exp, _ = spearmanr(x_exp, y_exp)
    kendall_corr_exp, _ = kendalltau(x_exp, y_exp)

    pearson_corr_u_shaped, _ = pearsonr(x_u_shaped, y_u_shaped)
    spearman_corr_u_shaped, _ = spearmanr(x_u_shaped, y_u_shaped)
    kendall_corr_u_shaped, _ = kendalltau(x_u_shaped, y_u_shaped)

    # Append the calculated correlations to the respective arrays for each relationship
    pearson_correlations_linear.append(pearson_corr_linear)
    spearman_correlations_linear.append(spearman_corr_linear)
    kendall_correlations_linear.append(kendall_corr_linear)

    pearson_correlations_exponential.append(pearson_corr_exp)
    spearman_correlations_exponential.append(spearman_corr_exp)
    kendall_correlations_exponential.append(kendall_corr_exp)

    pearson_correlations_u_shaped.append(pearson_corr_u_shaped)
    spearman_correlations_u_shaped.append(spearman_corr_u_shaped)
    kendall_correlations_u_shaped.append(kendall_corr_u_shaped)

# Step 4: Plot the results for each relationship
plt.figure(figsize=(15, 6))

# Linear Relationship
plt.subplot(131)
plt.plot(correlation_values, pearson_correlations_linear, label='Pearson', marker='o')
plt.plot(correlation_values, spearman_correlations_linear, label='Spearman', marker='s')
plt.plot(correlation_values, kendall_correlations_linear, label='Kendall', marker='x')
plt.xlabel('Correlation Coefficient (rho)')
plt.ylabel('Correlation Value')
plt.title('Linear Relationship')
plt.grid(True)

# Exponential Relationship
plt.subplot(132)
plt.plot(correlation_values, pearson_correlations_exponential, label='Pearson', marker='o')
plt.plot(correlation_values, spearman_correlations_exponential, label='Spearman', marker='s')
plt.plot(correlation_values, kendall_correlations_exponential, label='Kendall', marker='x')
plt.xlabel('Correlation Coefficient (rho)')
plt.ylabel('Correlation Value')
plt.title('Exponential Relationship')
plt.grid(True)

# U-shaped Relationship
plt.subplot(133)
plt.plot(correlation_values, pearson_correlations_u_shaped, label='Pearson', marker='o')
plt.plot(correlation_values, spearman_correlations_u_shaped, label='Spearman', marker='s')
plt.plot(correlation_values, kendall_correlations_u_shaped, label='Kendall', marker='x')
plt.xlabel('Correlation Coefficient (rho)')
plt.ylabel('Correlation Value')
plt.title('U-shaped Relationship')
plt.grid(True)

plt.tight_layout()
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import pearsonr, spearmanr, kendalltau
import matplotlib.pyplot as plt

# Step 1: Define the correlation values from -1 to 1 in steps of 0.05
correlation_values = np.arange(-1, 1.05, 0.05)

# Step 2: Create empty arrays to store correlation coefficients
sample_sizes = [10, 30, 100]

correlation_methods = ['Pearson', 'Spearman', 'Kendall']

# Initialize dictionaries to store the results
correlation_results = {method: {size: [] for size in sample_sizes} for method in correlation_methods}

# Step 3: Loop through correlation values, sample sizes, and relationships
for size in sample_sizes:
    for rho in correlation_values:
        # Create empty arrays to store correlation coefficients for this sample size
        pearson_correlations = []
        spearman_correlations = []
        kendall_correlations = []

        for _ in range(100):  # Repeat 100 times for more robust results
            # Linear Relationship
            np.random.seed(0)
            x = np.random.rand(size)
            y = rho * x + (1 - rho) * np.random.rand(size)

            # Calculate the correlation coefficients for each relationship
            pearson_corr, _ = pearsonr(x, y)
            spearman_corr, _ = spearmanr(x, y)
            kendall_corr, _ = kendalltau(x, y)

            # Append the calculated correlations to the respective arrays for each method
            pearson_correlations.append(pearson_corr)
            spearman_correlations.append(spearman_corr)
            kendall_correlations.append(kendall_corr)

        # Calculate the mean of correlation coefficients for each method and relationship
        mean_pearson_corr = np.mean(pearson_correlations)
        mean_spearman_corr = np.mean(spearman_correlations)
        mean_kendall_corr = np.mean(kendall_correlations)

        # Append the means to the respective dictionaries
        correlation_results['Pearson'][size].append(mean_pearson_corr)
        correlation_results['Spearman'][size].append(mean_spearman_corr)
        correlation_results['Kendall'][size].append(mean_kendall_corr)

# Step 4: Plot the results for each relationship, sample size, and correlation method
plt.figure(figsize=(18, 9))

for idx, method in enumerate(correlation_methods):
    for size in sample_sizes:
        plt.subplot(3, 3, idx + 1)
        plt.plot(correlation_values, correlation_results[method][size], label=f'n={size}')
        plt.title(f'{method} Correlation vs. Correlation Coefficient')
        plt.xlabel('Correlation Coefficient (rho)')
        plt.ylabel('Correlation Value')
        plt.grid(True)

plt.tight_layout()
plt.legend()
plt.show()
```
